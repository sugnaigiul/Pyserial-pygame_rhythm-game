# Analyse des données des joueurs
Ce notebook charge un fichier CSV contenant des données sur les joueurs et affiche trois visualisations distinctes :
- Les scores triés par ordre croissant
- Une relation entre le nombre de parties et les scores
- Une répartition des joueurs ayant joué une ou plusieurs parties

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Charger les données depuis le fichier CSV

```python
df = pd.read_csv('./../src/player_data.csv', header=None, names=['Player', 'Score', 'Games'])
df.head()
```

## Trier les données par Score pour le premier graphique

```python
df_sorted = df.sort_values(by='Score')
```

## Compter le nombre de joueurs avec une seule partie et plusieurs parties

```python
one_game_count = len(df[df['Games'] == 1])
multiple_games_count = len(df[df['Games'] > 1])

one_game_count, multiple_games_count
```

## Visualisation 1 : Scores triés en ordre croissant

```python

plt.figure(figsize=(8, 6))
plt.bar(df_sorted['Player'], df_sorted['Score'], color='skyblue')
plt.xlabel('Joueurs')
plt.ylabel('Highscore')
plt.title('Highscores des joueurs triés en ordre croissant')
plt.xticks(rotation=90)
plt.show()
```

Cette visualisation est interssante car elle nous permet d'observer la repartition des records dans la communauté des joueurs.

## Visualisation 2 : Relation entre le nombre de parties et les scores

```python

plt.figure(figsize=(8, 6))
plt.scatter(df['Games'], df['Score'], color='coral')
plt.xlabel('Nombre de parties')
plt.ylabel('Highscore')
plt.title('Highscore en fonction du nombre de parties')
plt.grid(True)
plt.show()
```

Cette visualisation nous permet d'observer une tendance dans laquelle les joueurs avec plusieurs parties score mieux que sont avec un faible nombre de parties.

## Visualisation 3 : Répartition des joueurs par nombre de parties

```python

plt.figure(figsize=(8, 6))
plt.pie(
    [one_game_count, multiple_games_count],
    labels=['1 partie', 'Plusieurs parties'],
    autopct='%1.1f%%',
    startangle=140,
    colors=['lightgreen', 'lightskyblue']
)
plt.title('Proportion de joueurs par nombre de parties')
plt.show()
```

Sur ce graphique, on peut observer la fidélité de la communauté de joueurs. On peut y observer qu'une partie des joueurs ne relance pas le jeu après avoir terminé pour la première fois.
